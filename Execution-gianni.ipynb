```python
%pylab inline
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
data = pd.read_csv("./DPIC Execution Database - U.S. Executions.csv ")
data
```

```python
gb_mean_note = data.groupby("Sex")["Number of Victims"].mean()
gb_mean_note
```

```python
data["QuadraKill or more"] = data["Number of Victims"] >= 3
gb_mean_KillStreak = data.groupby("Sex")["QuadraKill or more"].mean()
sns.barplot(x = gb_mean_KillStreak.index, y = gb_mean_KillStreak.values)
```

```python
gb_mean_juvie = data.groupby("Juvenile")["Number of Victims"].mean()
sns.barplot(x = gb_mean_juvie.index, y = gb_mean_juvie.values)
```

```python
gb_foreign_national = data.groupby("Foreign National")["Number of Victims"].mean()
sns.barplot(x = gb_foreign_national.index, y = gb_foreign_national.values)
```
